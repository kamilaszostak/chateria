## **Zadanie 1**
Które z następujących zadań wymagają w Twojej opinii inteligencji od człowieka:

*   wypełnianie deklaracji PIT,
*   **streszczanie tekstu,** -> *wymaga wysokiej inteligencji, szczególnie w zakresie rozumienia, interpretacji i wyciągania kluczowych informacji,*
*   **tłumaczenie tekstu** -> *wymaga znacznej inteligencji, szczególnie w rozumieniu kontekstu, niuansów językowych i kulturowych,*
*   klasyfikacja tekstu do kategorii tematycznych,
*   odpowiadanie na proste pytania zadawane w języku naturalnym (np. polskim),
*   układanie rozkładu jazdy transportu miejskiego,
*   programowanie (pisanie programów komputerowych) -> wymaga wysokiej inteligencji, w tym umiejętności analitycznego myślenia, rozwiązywania problemów i kreatywności.
*   „programowanie” kanałów telewizyjnych,
*   testowanie oprogramowania,
*   **komponowanie muzyki,** -> *Wymaga wysokiej inteligencji i kreatywności, ponieważ wymaga głębokiego rozumienia teorii muzyki i wyobraźni artystycznej,*
*   rozwiązywanie układów równań,
*   symboliczne obliczanie pochodnych funkcji,
*   symboliczne całkowanie funkcji,
*   **kierowanie samochodem.** -> *wymaga wysokiej inteligencji, zwłaszcza w zakresie przetwarzania informacji w czasie rzeczywistym, podejmowania decyzji i przewidywania sytuacji.*

Według mnie inteligencji od człowieka wymagają: streszczenie tekstu, tłumaczenie tekstu, komponowanie muzyki, kierowanie samochodem.

## **Zadanie 2**

Które z następujących problemów można uznać za mieszczące się w zakresie sztucznej inteligencji:

*   **streszczanie tekstu,** ->*(modele językowe AI są zdolne do streszczania tekstów, analizując treść i wyciągając kluczowe informacje)*
*   **tłumaczenie tekstu,** ->*(szczególnie zaawansowane systemy tłumaczeniowe (np. Google Translate), skutecznie tłumaczą teksty między językami, niestety nie zawsze AI rozumie kontekst wypowiedzi)*
*   **klasyfikacja tekstu do kategorii tematycznych,** -> *(algorytmy uczenia maszynowego są wykorzystywane do analizy tekstu i klasyfikacji danych)*
*   **odpowiadanie na proste pytania zadawane w języku naturalnym,** -> *(np. wirtualni asystenci, chat boty, system dialogowy)*
*   **rozwiązywanie układów równań,**->*(Mieści się w zakresie SI, jeśli dotyczy zastosowania algorytmów numerycznych i systemów ekspertowych do automatycznego rozwiązywania takich problemów)*
*   **układanie rozkładu jazdy,**->*(jest dziedziną SI, wymaga rozwiązywania problemów kombinatorycznych i czasem wykorzystania algorytmów heurystycznych)*
*   **rozwiązywanie układów równań liniowych,**-> *(zadanie mieści się w zakresie AI)*
*   **symboliczne obliczanie pochodnych,**->*(komputerowe systemy algebry symbolicznej jak Mathematica lub inne narzędzia wspierające symboliczne przekształcenia)*
*   **symboliczne całkowanie,**-> *(mieści się w zakresie SI, szczególnie w kontekście zastosowań w automatyzacji matematyki symbolicznej)*
*   **kierowanie samochodem.** ->*(to jedno z najbardziej zaawansowanych zastosowań SI, obejmujące wiele dziedzin, takich jak widzenie komputerowe, przetwarzanie sygnałów, podejmowanie decyzji, uczenie maszynowe, obecnie najczęściej jest stosowane jako wspomaganie jazdy, rzadziej jako zastępowanie człowieka jako kierowcy)*

Według mnie wszystkie z wymienionych puntków mieszczą się w zakresie SI.

##**Zadanie 3**
Które z poniższych rodzajów komunikacyjnego zachowania człowieka mogą być
obecnie skutecznie imitowane przez sztuczne systemy (odpowiednio oprogramowane
maszyny):

*   rozmowa towarzyska,-> *(obecnie chatboty, (np. GPT) potrafią imitować rozmowę towarzyską, ale brakuje im pełnego rozumienia kontekstu emocjonalnego i niuansów społecznych)*
*   dyskusja polityczna, -> *(możliwa do pewnego stopnia, szczególnie w zakresie przedstawiania faktów czy argumentów, ale systemy mają trudności z interpretacją złożonych opinii i emocji politycznych)*
*   dyskusja naukowa, -> *(Systemy mogą uczestniczyć w ograniczonych dyskusjach naukowych, zwłaszcza w dobrze zdefiniowanych obszarach wiedzy, ale nie potrafią proponować innowacyjnych teorii czy wniosków)*
*   odpowiadanie na pytania klientów w telefonicznej infolinii, -> *(możliwe, ale rozumienie mowy, emocji i niuansów pozostaje wyzwaniem dla SI)*
*   **odpowiadanie na pytania klientów w internetowej infolinii.** -> *(Bardzo skuteczne SI radzi sobie w zadaniach zadaniach dzięki chatbotom i systemom FAQ, które nie wymagają interpretacji tonu głosu ani emocji)*

Według mnie obecnie skutecznie SI radzi sobie imitując zachowania człowieka tylko w przypadku odpowiadania na pytania klientów w infolini internetowej.
